製作不同尺寸的圖像
有一個完整的Gridspec API (https://matplotlib.org/3.1.1/api/gridspec_api.html#module-matplotlib.gridspec)，我
們可以利用，但在這裡我們將採用一個捷徑，使用plt.subplots()的gridspec_kw參數。gridspec_kw是一個字典，
包含所有可以輸入Gridspec類物件的關鍵字參數。
在這裡，我們希望一張圖片比第二張細得多，所以我們將把width_ratios參數傳給gridspec_kw。這裡，我們設定
第二幅圖比第一幅圖寬3倍。


```python
fig, axes = plt.subplots(ncols=2, gridspec_kw={'width_ratios':[1,3]},constrained_layout=True)
#指定 constrained_layout=True在召唤 plt.subplots使布局受到适当的约束。

axes[0].plot(np.sort(data['Age']),'r',marker='o',markersize=1)
axes[0].set_title('Titanic Age')
axes[0].set_ylabel('Age')
axes[0].set_xlabel('people')
axes[0].set_yticks([15,25,55,64])
axes[0].set_yticklabels(['children','youth','adults','senior'])

axes[1].plot(np.sort(data['Fare']),marker='o',markersize=1)
axes[1].set_title('Titanic Fare')
axes[1].set_ylabel('Fare')
axes[1].set_xlabel('people')
axes[1].set_yscale('log') 
axes[1].set_yticks([3,12,30,870])
axes[1].set_yticklabels(['3rd','2nd','1st B','1st S'])

plt.show()
```

鐵達尼號資料集沒有連續的資料，故此處用正弦波來操作線圖。我們使用numpy來生成sine和cosine。這裡的x值
是100個在0和2𝜋之間等距的值 。

Make a line plot

```python
x = np.linspace(0, 2*np.pi, 100) # np.linspace 生成含浮點數的 array
sin_x = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,sin_x)
plt.show()
```

設置長寬比
正弦曲線上的x和y應該有相等的長寬，因為它們的單位是一樣的。為了反映這一點，我們使用了.set_aspect函
數。

```python
fig, ax = plt.subplots()
ax.plot(x, sin_x)
ax.set_aspect('equal')
plt.show()
```

Marker
使用marker kwarg在每個成對的x,y上放置一個標記。在標記參考marker reference
(https://matplotlib.org/examples/lines_bars_and_markers/marker_reference.html)中找到標記樣式的清單，或者
製作自訂標記custom markers
(https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/marker_reference.html)。使用figsize=(h,w) kwarg
製作圖形，這樣就可以看到標記。


```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x, sin_x, marker='*')  # 以*在每個成對的 x,y 上放置一個標記
ax.set_aspect('equal')
plt.show()
```

改變顏色
    color : 改變線的顏色
    markerfacecolor : 改變標記物的表面顏色
    markeredgecolor : 改變標記的邊緣顏色

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x, sin_x, marker='*', markerfacecolor = 'white', markeredgecolor = 'blue', color='red')
# 標記表面顏色改白色，標記外框藍色，線紅色

ax.set_aspect('equal')
plt.show()
```

改變尺寸

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x, sin_x, marker='*', markersize=20, linewidth=20)
# 線寬20，標記大小20

ax.set_aspect('equal')
plt.show()
```

透過添加材質來改變線條
有時我們需要改變線條的材質。linestyles parameter
(https://matplotlib.org/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markerslinestyles-py)

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x, sin_x, markersize=20, linestyle='-.')
# 線寬20，標記大小20

ax.set_aspect('equal')
plt.show()
```

有時候資料始以階梯間隔採樣，可以用 drawstyle 這格參數
(https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html?
highlight=drawstyle#matplotlib.lines.Line2D.set_drawstyle)


```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x, sin_x, drawstyle='steps')
ax.set_aspect('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x, sin_x, markersize=20, linestyle='-.', color='red', label='sin')
ax.set_aspect('equal')
ax.legend()
plt.show()
```

訂製顏色 colormap

    (https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html))。

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.scatter('Age','Fare',c='Survived',cmap='tab10',data=df)
ax.set(xlabel='age',ylabel='fare')
plt.show()
```

創建一個ListedColormap，以便我們可以手動設置顏色。我們從colors module
(https://matplotlib.org/api/colors_api.html)中獲得listed colormap
(https://matplotlib.org/api/_as_gen/matplotlib.colors.ListedColormap.html)。

```python
import matplotlib.colors as mcolors
cmap = mcolors.ListedColormap(['black','yellow'])
fig, ax = plt.subplots(figsize=(15,5))
ax.scatter('Age','Fare',c='Survived',cmap=cmap,data=df)
ax.set(xlabel='Age',ylabel='Fare')
plt.show()
```

添加圖例
    利用scatter legend
    (https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/scatter_with_legend.html#automated-legendcreation)功能，根據顏色生成一個標籤。這裡我們把ax.scatter的輸出指定給sc變數，因為我們需要使用sc物件來
    產生圖例。

```python
fig, ax = plt.subplots(figsize=(15,5))
cmap = mcolors.ListedColormap(['yellow','black'])
sc = ax.scatter('Age','Fare',c='Survived',cmap=cmap,data=df)
ax.set(xlabel='Age',ylabel='Fare')
ax.legend(*sc.legend_elements())
plt.show()
```

添加自訂標籤
sc.legend_elements()傳回兩個 list
    handle：圖例中顯示的繪圖物件（本例中為散點）
    labels：每個控制碼上的標籤 在本例中，標籤是0和1，因為那是作為顏色傳入的，所以我們用我們的自訂
            標籤覆蓋默認的標籤。


```python
fig, ax = plt.subplots(figsize=(15,5))
sc = ax.scatter('Age', 'Fare', c='Survived', cmap=cmap, data=df)
ax.set(xlabel = 'Age', ylabel = 'Fare')
ax.legend(handles = sc.legend_elements()[0], labels=['Died', 'Survived'])
plt.show()
```

另一種設置圖利的方法是對資料進行分割，並使用標籤函數

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.scatter('Age','Fare',c='black',cmap=cmap, data=df[df['Survived']==0], label='died')
ax.scatter('Age','Fare',c='Yellow',cmap=cmap, data=df[df['Survived']==1], label='Survived')
ax.set(xlabel='Age',ylabel='Fare')
ax.legend()
plt.show()
```

標記細部修改
    使用alpha keyword添加一些透明度，然後將點的邊界設為黑色，在資料量多時增加識別度

```python
fig, ax = plt.subplots(figsize=(15,5))
sc = ax.scatter('Age', 'Fare', c='Survived', cmap=cmap, alpha=.5, edgecolor='black', data=df)
ax.set(xlabel = 'Age', ylabel = 'Fare')
ax.legend(handles = sc.legend_elements()[0], labels=['Died', 'Survived'])
plt.show()
```

氣泡圖(Bubble Chart): 已呈現更多資訊
在年齡、票價和存活率以及乘客等級，是否存在某種模式? 再將乘客等及設定為另一格變數進行點大小的編碼

```python
fig, ax = plt.subplots(figsize=(15,5))
sc = ax.scatter('Age','Fare', c = 'Survived', s=df['Pclass']*25, cmap = cmap, alpha=0.5, edgecolor='black', data=df)
ax.set(xlabel='Age',ylabel='Color')
ax.legend(handles = sc.legend_elements()[0], labels=['Died','Survived'])
plt.show()
```
