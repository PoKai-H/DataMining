Concatenating & Append

Concatenating
    串接
    Combining together Series, DataFrame, and Panel objects
    pd.concat(objs, axis = 0, join='outer', join_axes=None, ingnore_index=False)
        objs - This is a sequence or mapping of Series, DataFrame, or Panel objects
        axis - {0, 1, ...}, default 0. This is the axis to concatenate along.
        join - {'inner', 'outer'}, default 'outer'. How to handle indexes on other axis(es)
                Outer for union and inner for intersection.
        ignore_index - boolean, default False. If True, do not use ther index values
                   on the concatenation axis. The resulting axis will be labeled 0, .., n-1.
        join_axes - This is the list of Index objects. Specific indexes to use for
                    the other (n-1) axes instead of performing inner/outer set logic


1. concat()
    1. 行、列對應
    2. 行、列不對應
        join = "outer" (聯集)預設
        join = "inner" (交集)

Merging & Joining

1. Merging
   The merge function allows you to merge DataFrames together using a similar logic as merging SQL tables together
        concat 用 list 連接、merge 直接寫在括號中
        merge 預設合併方向為橫向
        以 on = 表根據哪一欄合併
        以 how = 表 inner(交集) outer(聯集)、哪一 DataFrame為主

(1) 針對 (on) 一個相同欄位

```python
left = pd.DataFrame({"key":["K0","K1","K2","k3"],
                     "A":["A0","A1","A2","A3"],
                     "B":['B0',"B1","B2","B3"]  })
right = pd.DataFrame({"key":["K0","K1","K2","k3"],
                     "C":["C0","C1","C2","C3"],
                     "D":['D0',"D1","D2","D3"]  })
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')

# 針對 key 欄位合併 (此例 key 欄皆同，故直接合併)
# how 預設是 inner(交集)
```

```python
pd.concat([left,right],axis=1)

# 和 concat()比較
```

(2) 針對 (on) 數個不同欄位

```python
left = pd.DataFrame({"key1":["K0","K0","K1","k2"],
                     "key2":["K0","K1","K0","k1"],
                     "A":["A0","A1","A2","A3"],
                     "B":['B0',"B1","B2","B3"]  })
right = pd.DataFrame({"key1":["K0","K1","K1","k2"],
                      "key2":["K0","K0","K0","k0"],   
                      "C":["C0","C1","C2","C3"],
                      "D":['D0',"D1","D2","D3"]  })
```

```python
left
```

```python
right
```

```python
# 根據 key1, key2 同時考慮
pd.merge(left, right, on=['key1', 'key2'])

# how 預設為 inner 僅會列出交集部分
```

```python
# 根據 key1, key2 同時考慮
pd.merge(left, right, how='outer', on=['key1','key2'])

# how 設為 outer 取聯集
```

```python
pd.merge(left, right, how='right', on=['key1','key2'])

# how 設為 right，以 right 擁有之欄位為主 (類似 inner)
```

```python
pd.merge(left, right, how='left', on=['key1','key2'])

# how 設為 left，以 left 擁有之欄位為主 (類似 inner)
```

2. Joining

```python
left = pd.DataFrame({'A': ['A0','A1','A2'],
                     'B': ['B0','B1','B2']},
                     index=['K0','K1','K2']
                    )
right = pd.DataFrame({'C': ['C0','C2','C3'],
                      'D': ['D0','D2','D3']},
                      index=['K0','K2','K3']
                    )
```

```python
left
```

```python
right
```

```python
left.join(right)

# 根據 left 擁有的欄位加上 right
```

```python
right.join(left, how='outer')

# 根據 right 擁有的欄位加上left , how='outer'取聯集, 故所有欄位都會列出
```

8. Pivot Table

```python
data = {'A':['foo','foo','foo','bar','bar','bar'],
        'B':['one','one','two','two','one','one'],
        'C':['x','y','x','y','x','y'],
        'D':[1,3,2,5,4,1]}

df = pd.DataFrame(data)
df
```

```python
df.pivot_table(values='D', index='A',columns=['C'])

# 用 A 做 index, C 做 columns, 取 D 的值計算平均
```

```python
df.pivot_table(values='D', index='A', columns=['C'], aggfunc='sum')

# 用 A 做 index, C 做 columns, 取 D 的值計算總和
# sum std count max min 亦可
```

```python
df.pivot_table(values='D',index=['A','C'],columns=['C'])
```
