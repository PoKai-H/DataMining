plt.subplot()
用於直接指定劃分方式和位置進行繪圖
    第一個參數: 子圖的列數
    第二個參數: 子圖的行數
    第三個參數: 每行的第幾個圖像

```python
plt.subplot(2,2,1)  
plt.subplot(2,2,2)
plt.subplot(2,2,3)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--') # r-- 紅色虛線
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title A')

plt.subplot(2,1,2)
plt.plot(x,y,'g--') # g-- 綠色虛線
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title B')
```

Matplotlib Object Directed Method

```python
# 創建一個空畫布
fig = plt.figure() 

# 增加座標軸組
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # x軸座標、y軸座標、寬、高，(0,0)原點對應 canvas 的位置

# 在坐標軸上畫圖
axes.plot(x,y,'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axesA = fig.add_axes([0.1,0.1,0.8,0.8])
axesB = fig.add_axes([0.2,0.5,0.4,0.3])

axesA.plot(x,y,'b')
axesA.set_xlabel('X Label')
axesA.set_ylabel('Y Label')
axesA.set_title('Title A')

axesB.plot(y,x,'r')
axesB.set_xlabel('X Label')
axesB.set_ylabel('Y Label')
axesB.set_title('Title B')
```

subplots()
    a more automatic axis mamager

```python
#　類似 plt.figure(), 但使用 tuple 來取出圖和軸
fig, axes = plt.subplots()

# 添加內容
axes.plot(x,y,'b')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

可在創建 subplots() 物件時指定行與列數量

```python
# 1*2 個子圖的空白畫布
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
# Axes 是一個用於繪圖的座標軸陣列
axes
```

```python
# 可通過此陣列進行反覆運算
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
```

matplotlib 一個常見的問題是子圖或數字重疊，可透過fig.tigt_layout() or plt.tight_layout方法
自動調整

```python
fig, axes = plt.subplots(nrows=3,ncols=3)

axes[0][1].plot(x,y)
axes[0][1].set_title('First Plot')
axes[0][1].set_xlabel('X Label')
axes[0][1].set_ylabel('Y Label')

axes[1][1].plot(y,x)
axes[1][1].set_title('Second Plot')
axes[1][1].set_xlabel('X Label')
axes[1][1].set_ylabel('Y Label')

plt.tight_layout()
```

Legends
    1. Use the label='label text' keyword argument when plots or other objects are added to the figure
    2. Use the legend method without arguments to add the legend to the figure
    多條線畫在同一張圖上 ax.legend()

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.plot(x, 2*x, label='2X')
ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')

ax.legend() #多條線畫在同一張圖上
```

3.The location of Legend
    loc that can be used to specify where in the figure the legend is to be drawn

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower right corner
ax.legend(loc=4) # lower left corner

# Most common to choose
ax.legend(loc=0) # Let matplotlib decide the optimal place
fig
```

Setting colors, linewidths, linetypes

Colors with MatLab like syntax
    the MATLAB-like syntax where 'b' means blue, 'g' means green, etc.
    selecting line styles: where, for example, 'b.-' means a blue line with dots

```python
# MATLAB style line color and style
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
```

Colors with the color = parameter
    define colors by their names or RGB hex codes
    Alpha indicates opacity

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue', alpha=0.5) # alpha=0.5 半透明 (half-transparant)
ax.plot(x, x+2, color='#8D008B') # RGB hex code
ax.plot(x, x+3, color='#FF8C00') # RGB hex code
```

Plot range
We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3) # 這裡這樣寫前面兩個是第一條線的參數，後兩個是第二條線的參數
axes[0].set_title('default axes ranges')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('custom axes range')
```

Histograms

```python
# Generate a normal distrubution, mean and std are 100 and 15 respectively
IQ = np.random.normal(100, 15, 100000)
plt.hist(IQ, density=True, bins=100)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(40, .025, r'$\mu=100,\ \sigma=15$')
plt.show()
```

```python
# Generate a normal distribution, center at 100 and y=150
x1 = np.random.normal(100, 15, 100000)  # np.random.normal (均值, 標準差, 樣本數)
x2 = np.random.normal(100, 30, 100000)

fig, axes = plt.subplots(1, 2, figsize=(10,5), tight_layout=True)

#We ccan set the number of bins with the `bins` kwarg

axes[0].hist(x1, bins=100, color='g')
axes[0].set_title('Histogram of IQ at mean = 100')
axes[0].set_xlabel('IQ')
axes[0].set_ylabel("'Probability'")

axes[1].hist(x2, bins=100)
axes[1].set_title('Historgram of IQ at mean = 150')
axes[1].set_xlabel('IQ')
axes[1].set_ylabel("'Probability'")

plt.show()
```
